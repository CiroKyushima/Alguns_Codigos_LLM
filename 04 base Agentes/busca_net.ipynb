```python
from duckduckgo_search import DDGS
from openai import OpenAI
```

```python
client = OpenAI(api_key="")
```

```python
def busca_web(pergunta):
    with DDGS() as ddgs:
        resultados = ddgs.text(pergunta, max_results=3)
        textos = [r["body"]for r in resultados]
        return "\n".join(textos)
```

```python
def agente(pergunta):
    info = busca_web()
    prompt = f'''voce é um agente inteligente. o usuario fez a seguinte pergunta: {pergunta}.
                 voce encontrou as seguinte respostas: {info}
                 responda apenas com base na informação acima
                 se não houver dados suficiente, responda que voce não conseguiu respostas'''
    
    resposta = client.chat.completions.create(
        model = "gpt-3.5-turbo",
        messages = [{"role": "user", "content": prompt}]
    )

    return resposta.choices[0].message.content
```

```python
resposta = agente()
print(resposta)
```
