um codigo simples ensinando a usar um modelo de LLM para geração textual

```python
from transformers import pipeline
from transformers import logging

logging.set_verbosity_error()
```

```python
gerador = pipeline("text-generation", model="pierreguillou/gpt2-small-portuguese")
text = "em sentido estrito, ciencia refere-se ao sistema de adquirir conhecimento baseado no método cientifico"
resultado = gerador(text, max_new_tokens=100)

print(resultado)
```
