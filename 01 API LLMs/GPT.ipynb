codigo simples para utilizar a API do GPT 

```python
from openai import OpenAI
```

```python
client = OpenAI(api_key="a key aqui :)  ")
model = "gpt-3.5-turbo"
```

```python
response = client.chat.completions.create(
    model = model,
    message = [
        {"role":"assistant","content": "tell me a joke"}
    ]
)
```

```python
message  = response.choice[0].message.content
print(message)
```
