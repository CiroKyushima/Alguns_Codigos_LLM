codigo simples para utilizar a API do GEMINI

```python
import google.generativeai as genai
```

```python
genai.configure(api_key="bote a key here ")
```

```python
model = genai.GenerativeModel("gemini-1.5-pro")
conversation = "voce é um assistente util que responde de maneira clara e objetiva" + "cite uma frase de carl sagan"
response = model.generate_content(conversation)
```

```python
print(response.text.strip())
```
