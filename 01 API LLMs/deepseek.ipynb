apenas um codigo simples de como utilizar o deepseek

```python
from openai import OpenAI
url = "https://api.deepseek.com"
api_key = "bota key, tira key, bota key"
```

```python
client = OpenAI(api_key=api_key, base_url=url)
```

```python
response = client.chat.completions.create(
    model="deepseek-chat",
    message={
        {"role": "system", "content:" : "voce é um assistente multi idioma que deve responder de forma clara e objetiva"},
        {"role": "user" , "content:" : "cite uma frase de carl sagan"}
    }
)
```

```python
print(response.choices[0].message.content)
```

abaixo outra forma de utlizar o deepseek

```python
import requests
import json
```

```python
url = "https://api.deepseek.com/chat/completions"
```

```python
headers = {
    "content-Type"  : "application/json",
    "Authorization" : f"Bearer {api_key}"
}

payload = {
    "model" : "deepseek-chat",
    "messages" : [
        {"role": "system", "content": "voce é um assistente que deve responder de maneira clara e objetiva"},
        {"role": "user", "content": "cite uma frase de carl sagan"}
    ]
}
```

```python
response = requests.post(url, headers=headers, data=json.dumps(payload))
result = response.json()
```

```python
print(result['choice'][0]['message']['content'])
```
