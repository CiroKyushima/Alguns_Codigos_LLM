codigo simples para geração de imagens com o modelo stable diffusion<br>
esse modelo é open source, e um pouco pesado

```python
from diffusers import StableDiffusionXLPipeline
import torch
import matplotlib.pyplot as plt
```

```python
pretrained_model = "stabilityai/stable-diffusion-xl-base-1.0"
prompt = "a colorful vibrant city street, daylight, a red car parked by the sidewalk"
negative_prompt = "blurry, low quality,  grayscale, monochrome, dull, dark"
num_inference_steps = 30
guidance_scale = 7.5
height = 768
width = 768
seed = 42
```

```python
generator = torch.manual_seed(seed)
pipeline = StableDiffusionXLPipeline.from_pretrained(
    pretrained_model,
    torch_dtype=torch.float16,
    use_safetensors=True
).to("cuda")
```

```python
image = pipeline(
prompt = prompt,
negative_prompt = negative_prompt,
height=height,
width=width,
num_inference_steps=num_inference_steps,
guidance_scale=guidance_scale,
generator=generator
).image[0]
```

```python
plt.imshow(image)
plt.axis('off')
plt.show()
```
