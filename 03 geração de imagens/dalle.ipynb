apenas um codigo simples para utilizar o DALL-E

```python
from openai import OpenAI
from PIL import Image
import base64
import requests
from io import BytesIO
```

```python
client = OpenAI(api_key="bota key, bota vai, vai me negar uma key?")
```

```python
response = client.images.generate(
    model = 'dall-e-3',
    prompt = "centro de controle futurista com hologramas e cista para uma cidade cyberpunk",
    size = "1024,1024",
    quality = "standard",
    n = 1
)
```

```python
image_url = response.data[0].url
print(image_url)
```

```python
image_response = requests.get(image_url)
img = Image.open(BytesIO(image_response.content))
img.show()
```
